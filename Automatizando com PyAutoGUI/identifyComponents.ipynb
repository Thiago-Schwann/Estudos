```python
import pyautogui

#Desse modo, podemos encontram um elemento específico na tela sem coordenadas exatas. A saída é exatamente a coordenada da imagem que passamos no parâmetro.
local = pyautogui.locateOnScreen('imagem.png')
print(local) 

#Com a saída na variável local, é possível usala como parâmetro para interação
pyautogui.moveTo(local)
pyautogui.click()
```

```python
import pyautogui

#Ao usar o confidence, aumentamos a precisão do pyautoGUI de encontrar imagens, minimizando erros. 
local = pyautogui.locateOnScreen('imagem.png', confidence=0.9)

```

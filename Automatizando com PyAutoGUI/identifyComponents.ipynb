```python
import pyautogui

#Desse modo, podemos encontram um elemento especfico na tela sem coordenadas exatas. A sada  exatamente a coordenada da imagem que passamos no parmetro.
local = pyautogui.locateOnScreen('imagem.png')
print(local) 

#Com a sada na varivel local,  possvel usala como parmetro para interao
pyautogui.moveTo(local)
pyautogui.click()
```

```python
import pyautogui

#Ao usar o confidence, aumentamos a preciso do pyautoGUI de encontrar imagens, minimizando erros. 
local = pyautogui.locateOnScreen('imagem.png', confidence=0.9)

```
